```python
df=pd.read_csv("CarPrice_Assignment.csv")
```

```python
df
```

```python
inputs=df.drop(['CarName'],axis='columns')
```

```python
target=df.price
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()
```

```python
inputs['fueltype']=Numerics.fit_transform(inputs['fueltype'])
inputs
```

```python
inputs_n=inputs.drop(['fueltype','price'],axis='columns')
inputs_n
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(inputs_n,target)
```

```python
pred=model.predict([[2013,430000,1,1]])
print(pred)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
training_data_prediction = model.predict(inputs_n)
```

```python
from sklearn import metrics
```

```python
error_score = metrics.r2_score(target, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(target, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show
```
